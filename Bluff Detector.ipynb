##**Simple Linear Regression**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn import linear_model
```

```python
df=pd.read_csv('canada_pci.csv')
```

```python
df
```

```python
x=df.iloc[:,0:1].values
x
```

```python
y=df.iloc[:,1].values
y
```

```python
from sklearn import linear_model
```

```python
lin_reg=linear_model.LinearRegression()
```

```python
lin_reg.fit(x,y)
```

```python
lin_reg.predict(x)
```

```python
pred_values=lin_reg.predict(x)
```

```python
df['predicted values']=lin_reg.predict(x)
```

```python
df.head()
```

```python
# plot actual values
plt.plot(x,y,color='green',label='Actual Values')
plt.scatter(x,y,color='green')
# plot predicted values
plt.plot(x,pred_values,color='red',label='Predicted Values')
plt.scatter(x,pred_values,color='red')
plt.title('years vs pci')
plt.xlabel('years')
plt.ylabel('pci')
plt.legend()
plt.show()
```

```python
from sklearn.metrics import r2_score

# Assuming y_true are the true values and y_pred are the predicted values
r_squared = r2_score(y, pred_values)
r_squared
```

##**Polynomial Regression**

```python
#Degree 2
from sklearn.preprocessing import PolynomialFeatures
poly_reg_2 = PolynomialFeatures(degree=2)
x_poly_2 = poly_reg_2.fit_transform(x)
x_poly_2
```

```python
lin_reg.fit(x_poly_2,y)
```

```python
lin_reg.predict(x_poly_2)
```

```python
pred_values=lin_reg.predict(x_poly_2)
```

```python
df['predicted values']=pred_values
```

```python
df
```

```python
# Plot Actual Values
plt.scatter(x,y,color='blue')
plt.plot(x,pred_values,color='red',label='Regression Line')
plt.title('Model Using Polynomial Regression Degree = 2')
plt.xlabel('Years')
plt.ylabel('PCI')
plt.legend()
plt.show()
```

```python
from sklearn.metrics import r2_score

# Assuming y_true are the true values and y_pred are the predicted values
r_squared = r2_score(y, pred_values)
r_squared
```
